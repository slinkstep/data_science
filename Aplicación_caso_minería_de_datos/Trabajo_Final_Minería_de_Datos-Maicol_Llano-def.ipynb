# 5. Feature Engineering

## 5.1 Escalamiento - Transformación

```python
df['skill_moves']=df['skill_moves']*20 # Se escala esta variable para que quede entre 0 y 100
df['weak_foot']=df['weak_foot']*20     # Se escala esta variable para que quede entre 0 y 100

# Conversión de categórico a numérico

var=['work_rate_att','work_rate_def']

for i in var:
    df[i].replace('Low', 0, inplace=True)
    df[i].replace('Medium', 50, inplace=True)
    df[i].replace('High', 100, inplace=True)
    
df['preferred_foot'].replace('Left', 0, inplace=True)
df['preferred_foot'].replace('Right', 1, inplace=True)
    
```

```python
# Conversión de booleano a numérico

boolean=df.dtypes[df.dtypes=='bool'].index.values
       
for i in boolean:
    df[i].replace(False, 0, inplace=True)
    df[i].replace(True, 1, inplace=True)
```

##  5.2 Creación de Variables

```python
# Se asume que la posición de un jugador es una de sus posiciones preferidas

preferedpos=['prefers_rs', 'prefers_rw', 'prefers_rf', 'prefers_ram',
'prefers_rcm', 'prefers_rm', 'prefers_rdm', 'prefers_rcb',
       'prefers_rb', 'prefers_rwb', 'prefers_st', 'prefers_lw',
       'prefers_cf', 'prefers_cam', 'prefers_cm', 'prefers_lm',
       'prefers_cdm', 'prefers_cb', 'prefers_lb', 'prefers_lwb',
       'prefers_ls', 'prefers_lf', 'prefers_lam', 'prefers_lcm',
       'prefers_ldm', 'prefers_lcb', 'prefers_gk']

pos=['rs', 'rw', 'rf', 'ram',
'rcm', 'rm', 'rdm', 'rcb',
       'rb', 'rwb', 'st', 'lw',
       'cf', 'cam', 'cm', 'lm',
       'cdm', 'cb', 'lb', 'lwb',
       'ls', 'lf', 'lam', 'lcm',
       'ldm', 'lcb', 'gk']

posiciones=[]

for i in range(len(df)):
    contador=0
    for j in range(len(preferedpos)):
        
        if df.iloc[i][preferedpos[contador]]==True:
            
            posiciones.append(pos[contador])
            break 
        
        contador=contador+1
```

```python
# Se pasa la variable de posición a numérica teniendo en cuenta la siguiente distribución en el campo

# 0  --> Arquero
# 1  --> Defensas
# 2  --> Mediocampistas
# 3  --> Ofensivos

df['posicion']=posiciones
df['pos']=posiciones

        
df['pos'].replace('gk', 0, inplace=True)
df['pos'].replace(['lb','cb','rb'] , 1, inplace=True)
df['pos'].replace(['lwb','cdm','rwb','lm','cam','cm','rm'] , 2, inplace=True)
#df['pos'].replace(['lm','cam','cm','rm'] , 3, inplace=True)
df['pos'].replace(['lw','st','rw','cf'] , 3, inplace=True)
```

```python
import seaborn as sns
#sns.set(style="darkgrid")

plt.figure(figsize=(15,8))
plt.title('Posiciones [cat]')
ax = sns.countplot(x='posicion', data=df)

for p in ax.patches: 
    if np.isnan(p.get_height()):
        height=0
    else: 
         height = p.get_height() 
            
    ax.text(p.get_x()+p.get_width()/2., 
      height + 3, 
      int(height), 
      ha="center") 
```

```python
from sklearn.cross_validation import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report,confusion_matrix


test_size = .30
seedNo = 42

x_train,X_test,y_train,Y_test = train_test_split(xx,y,test_size = test_size, random_state = seedNo)


# Choose algorithm and train
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.neural_network import MLPClassifier
from sklearn.ensemble import AdaBoostClassifier
import xgboost as xgb


from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score

mymodel = []
mymodel.append(('LogReg', LogisticRegression()))
mymodel.append(('SVM', SVC()))
mymodel.append(('DeciTree', DecisionTreeClassifier()))
mymodel.append(('RandForest', RandomForestClassifier()))
mymodel.append(('KNN', KNeighborsClassifier()))
mymodel.append(('NaiveBayes', GaussianNB()))
mymodel.append(('Neuronal', MLPClassifier()))
mymodel.append(('XGBClassifier', xgb.XGBClassifier()))
mymodel.append(('AdaBoostClassifier', AdaBoostClassifier()))



resultados = {}
cm=[]
for algoname, algorithm in mymodel:    
    
    
    algorithm.fit(x_train,y_train)
    pred = algorithm.predict(X_test)
    
    Eval_result=accuracy_score(Y_test, pred)
    resultados[algoname]= Eval_result
    
    print("Modelo y precisión:", algoname, 100*Eval_result,"%")
    print("")
    print(classification_report(Y_test, pred))
    print("--------------------------------------------------------------")
    cmtemp=confusion_matrix(Y_test, pred)
    cm.append(cmtemp / cmtemp.astype(np.float).sum(axis=1,keepdims=True))
```
