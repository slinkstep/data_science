# 6. Detección de factores críticos

Se realizará un análisis estadístico entre las variables de entrada y nuestro target, con el objetivo de determinar cuáles están más relacionadas, es decir que tienen una dependencia lineal marcada. (Para llevar a cabo este análisis se usará la función SelectKBest que realiza los análisis basada en una función de test que como se dijo anteriormente estimaelgrado dedependencia lineal entre dos variables aleatorias).

## 6.1 Variable JobSatisfaction

```python
from sklearn.feature_selection import SelectKBest

# Funcion para identificar mejores variables basadas en pruebas estadísticas univariadas
def Feature_Selection(df,target):
    select = SelectKBest()
    select.fit(df.drop([target],axis=1), df[target])
    scores = select.scores_
    feature_scores = pd.DataFrame({'columnas': df.drop([target],axis=1).columns.tolist(),'scores': scores.tolist(),})
    return feature_scores.sort_values(by='scores',ascending=False)
```

```python
# Funcion para identificar mejores variables basadas en pruebas estadísticas univariadas
Feature_Selection(encoded_df,'JobSatisfaction')
```

###### Se observa que

- Una de las variables que más se relaciona con nuestro target es la de Atrittion, la cual indica si una persona dejó la compañía, de acuerdo con esto, esta variable será tenida en cuenta para análisis futuros con el objetivo de identificar factores que llevan a que una persona deje la compañía por el hecho de estar insatisfecho con su trabajo. Esta relación se puede entender fácilmente ya que es muy probable que si una persona está inconforme con su trabajo abandonde la compañía.

```python
# Se obtiene el TOP 10 de factores que influyen con la satisfacción laboral

Feature_Selection(encoded_df,'JobSatisfaction').head(10)['columnas'].values
```

# 9. Algoritmos de clasificación para predecir la decersión laboral

## 9.1 Preparación del Dataset

```python
from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import StandardScaler

y=encoded_df['Attrition']
xx=encoded_df.drop('Attrition',axis=1)
ScaleFn = MinMaxScaler()
xx = ScaleFn.fit_transform(xx)

# Se aplica un escalamiento a los datos, para tener un mejor desempeño a la hora de aplicar los algoritmo de
# clasificación
```

## 9.2 Aplicación de modelos

```python
from sklearn.cross_validation import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report,confusion_matrix
import warnings; warnings.simplefilter('ignore')

# Se divide el dataset con datos de prueba y datos de test

test_size = .30
seedNo = 42

x_train,X_test,y_train,Y_test = train_test_split(xx,y,test_size = test_size, random_state = seedNo)



# Choose algorithm and train
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.neural_network import MLPClassifier
from sklearn.ensemble import AdaBoostClassifier
import xgboost as xgb
from sklearn.ensemble import GradientBoostingClassifier


from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score

mymodel = []
mymodel.append(('LogReg', LogisticRegression()))
mymodel.append(('SVM', SVC()))
mymodel.append(('DeciTree', DecisionTreeClassifier()))
mymodel.append(('RandForest', RandomForestClassifier()))
mymodel.append(('KNN', KNeighborsClassifier()))
mymodel.append(('NaiveBayes', GaussianNB()))
mymodel.append(('Neuronal', MLPClassifier()))
mymodel.append(('XGBClassifier', xgb.XGBClassifier()))
mymodel.append(('AdaBoostClassifier', AdaBoostClassifier()))
mymodel.append(('GradientBoostClassifier', GradientBoostingClassifier()))





resultados = {}
cm=[]
for algoname, algorithm in mymodel:    
    
    
    algorithm.fit(x_train,y_train)
    pred = algorithm.predict(X_test)
    
    Eval_result=accuracy_score(Y_test, pred)
    resultados[algoname]= Eval_result
    
    print("Modelo y precisión:", algoname, 100*Eval_result,"%")
    print("")
    print(classification_report(Y_test, pred))
    print("--------------------------------------------------------------")
    cmtemp=confusion_matrix(Y_test, pred)
    cm.append(cmtemp / cmtemp.astype(np.float).sum(axis=1,keepdims=True))
```

## 9.4 Cross-Validation

```python
import warnings; warnings.simplefilter('ignore')
seedNo = 42

y=encoded_df['Attrition']
xx=encoded_df.drop('Attrition',axis=1)
ScaleFn = MinMaxScaler()
xx = ScaleFn.fit_transform(xx)

# Choose algorithm and train
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.neural_network import MLPClassifier
from sklearn.ensemble import AdaBoostClassifier
import xgboost as xgb
from sklearn.ensemble import GradientBoostingClassifier


from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score

mymodel = []
mymodel.append(('LogReg', LogisticRegression()))
mymodel.append(('SVM', SVC()))
mymodel.append(('DeciTree', DecisionTreeClassifier()))
mymodel.append(('RandForest', RandomForestClassifier()))
mymodel.append(('KNN', KNeighborsClassifier()))
mymodel.append(('NaiveBayes', GaussianNB()))
mymodel.append(('Neuronal', MLPClassifier()))
mymodel.append(('XGBClassifier', xgb.XGBClassifier()))
mymodel.append(('AdaBoostClassifier', AdaBoostClassifier()))
mymodel.append(('GradientBoostClassifier', GradientBoostingClassifier()))



resultados={}

for algoname, algorithm in mymodel:    
    kfoldFn = KFold(n_splits = 10, random_state = seedNo)
    Eval_result = cross_val_score(algorithm, xx, y, cv = kfoldFn, scoring = 'accuracy')  
    resultados[algoname]=Eval_result.mean()

    print("Modelo y precisión:", algoname, 100*Eval_result.mean(),"%")
```
